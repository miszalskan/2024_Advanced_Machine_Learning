### Iterative Reweighted Least Squares

```python
import numpy as np
```

```python
def logit(x):
  def logit_el(x):
    return np.log(x / (1 - x))
  if isinstance(x, list):
    x = np.array(x)
  if isinstance(x, float) or isinstance(x, int)  or isinstance(x, np.ndarray):
    return logit_el(x)
  else:
    return 
```

```python
def logistic(x):
    def logistic_el(x):
        return 1/(1+np.exp(-x))
    if isinstance(x, list):
        x = np.array(x)
    if isinstance(x, float) or isinstance(x, int)  or isinstance(x, np.ndarray):
        return logistic_el(x)
    else:
        return
```

```python
def var(x):
    def v_el(x):
        return x*(1-x)
    if isinstance(x, list):
        x = np.array(x)
    if isinstance(x, float) or isinstance(x, int)  or isinstance(x, np.ndarray):
        return v_el(x)
    else :
        return
```

```python
def logit_prime(x):
    def logit_prime_el(x):
        return 1/(x*(1-x))
    if isinstance(x, list):
        x = np.array(x)
    if isinstance(x, float) or isinstance(x, int)  or isinstance(x, np.ndarray):
        return logit_prime_el(x)
    else:
        return
```

```python
import scipy
```

```python
def WLS(X, W, z):
    return scipy.linalg.inv(X.T.dot(W).dot(X)).dot(X.T.dot(W).dot(z))

```

```python
def ll(p, y):
    return sum(y*np.log(p)+(1-y)*np.log(1-p))
```

```python
def IRLS(X,y):
    mu = [0.5]*len(y)
    delta = 1
    i = 1
    LL = 0 

    while delta> 0.00001 :
        Z = logit(mu) + (y-mu)*logit_prime(mu)
        W =  np.diag((1 / (logit_prime(mu)**2 * var(mu))))
        beta = WLS(X = X, W = W, z = Z)
        eta = np.array(X @ beta)
        mu = logistic(eta)
        LL_old = LL
        LL = ll(p = mu, y = y)
        delta = np.abs(LL - LL_old)

        print("Iteration: "+ str(i)+ " LL: "+ str(LL) + " Beta: "+str(beta) )
        i += 1 
    return beta      


```
